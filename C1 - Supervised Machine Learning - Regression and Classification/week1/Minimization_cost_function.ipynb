```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#The model is f=wx
#The training set is:
x_train = np.array([1,2,3])
y_train = np.array([1,2,3])

plt.figure(figsize=(5, 3.5))
plt.scatter(x_train,y_train, marker = 'x', c = "r")
plt.title("Training set")
plt.show()
```

We can only vary the argument (w), therefore we can define a range of numbers to change the argument. For example, between 0 and 2 and calculate the cost function 

```python
#Function that calulates the cost functiom
def cost_fun(w,x_train,y_train):
    m = len(x_train)
    j = 0
    for i in range(0,m):
        j+=(w*x_train[i]-y_train[i])**2
    J = j/(2*m)
    return J
```

```python
#Define the values
min_val = 0
max_val = 2
w_values= np.arange(min_val,max_val,0.01)
values = len(w_values)

#Caluclate this cost function for every w value:
Costs = []
for i in w_values:
    Costs.append(cost_fun(i,x_train,y_train))
```

```python
#Plot it:
plt.scatter(w_values,Costs,c='b',s=5)
plt.xlabel("Value of w")
plt.ylabel("Value of J")
plt.title(f"J depending on w between {min_val} and {max_val}")
plt.show()
```

We can clearly see that the cost function gets lower near the w=1, but if we were`t that sure, we could find the minimum value 

```python
print(f"The minimum value of the Cost Function is: {min(Costs)}")
```
